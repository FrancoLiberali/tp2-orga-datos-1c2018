```python
df_tmp = pd.merge(df, avisos.df[list(FEATURES)], left_on='idaviso', right_index=True, how='left')
```

```python
vistas_data = pd.merge(vistas.df, avisos.df[list(FEATURES)], left_on='idaviso', right_index=True, how='inner')
```

```python
gb = vistas_data.groupby('idpostulante')['nombre_area'].value_counts()
```

```python
df_tmp['cant_vistas_area'] = df_tmp['idpostulante'] + '-' + df_tmp['nombre_area']
```

```python
def get_cant(x):
    postu, *area = x.split('-')
    clave = (postu, '-'.join(area))
    if clave in gb.index:
        return gb[clave]
    return 0
df_tmp['cant_vistas_area'] = df_tmp['cant_vistas_area'].map(get_cant)
```

```python
df_tmp[df_tmp['cant_vistas_area'] > 1]
```
